```python
import pyttsx3
import datetime
import speech_recognition as sr
import wikipedia
import webbrowser
import os
engine = pyttsx3.init('sapi5')
voices = engine.getProperty('voices')
#print(voice[0].id)

engine.setProperty('voice',voices[0].id)




def  speak(audio):
    engine.say(audio)
    engine.runAndWait()
    
    
def wishme():
    hour = int(datetime.datetime.now().hour)
    if hour>=0 and hour<12:
        speak("good morning!")
        
    elif hour>=12 and hour<18:
            speak("good after noon!")
            
            
    else:
        speak("goood evening!")
    speak("i am jarvis 2.0. please tell me how would i help you ")
    
def takecommand():
    r= sr.Recognizer()
    with sr.Microphone() as source:
            print("listening.....")
            r.pause_threshold = 0.1
            audio = r.listen(source)
            
    try:
        print("Recognizing....")
        query = r.recognize_google(audio, Language='en_in')
        print(f"user said: {query}\n")
                  
    except Exception as e:
        #print(e)
        print("say that again please")
        return "none"
                  
    return query
                  
            
        


if __name__ == "__main__":
    wishme()
    while True:
        query = takecommand().lower()
        
        
        
        #logic on exicuting task  based n query
        
        if 'wikipedia' in query:
            speak('searching wikipedia.......')
            query = query.replace("wikipedia","")
            result = wikipedia.summary(query,sentences = 2)
            speak("acording to wikipedia")
            speak(result)

        elif 'open youtube'in query:
            webbrowser.open("youtube.com")


        elif 'open google' in query:
            webbrowser.open("google.com")

        elif 'stackoverflow' in query:
            webbrowser.open("stackoverflow.com")

        elif 'play music' in query:
            music_dir = 'C:\\Users\\HRITIK\\Downloads\\SHAREit\\mp3 you tube'
            songs = os.listdir(music_dir)
            print(songs)
            os.startfile(os.path.join(music_dir,songs[0]))
            
```
