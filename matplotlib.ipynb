```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0.0,2.0,0.01)
s = 1+np.cos(2*np.pi*t)

plt.plot(t,s,'--')
plt.grid()
plt.xlabel('time')
plt.ylabel('voltage')
plt.title('cos wave graph ')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0.0,2.0,1.0)
s = 1+np.cos(2*np.pi*t)

plt.plot(t,s)
plt.xlabel('time')
plt.ylabel('voltage')
plt.title('sin graph')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1=np.linspace(0.0,5.0)
x2=np.linspace(0.0,2.0)

y1=np.cos(2*np.pi*x1)*np.exp(-x1)
y2=np.cos(2*np.pi*x1)

plt.subplot(2,1,1)
plt.plot(x2,y2,'o-')
plt.title('subplot-1')
plt.xlabel('x1')
plt.ylabel('Amp(y1)')

plt.subplot(2,1,2)
plt.plot(x2,y2,'.-')
plt.title('subplot-2')
plt.xlabel('x2')
plt.ylabel('Amp(y2)')
plt.show()
```
